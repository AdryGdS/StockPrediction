```python
# Criar a coluna de target: se o fechamento de amanh for maior que o de hoje
base['target'] = base['Close'][1:len(base)].reset_index(drop=True)
base.tail()
```

```python
# Preparar os dados para previso (remover a coluna target da ltima linha)
prev = base[-1::].drop('target', axis=1)
```

```python
# Separar os dados de treino (remover a ltima linha de base)
treino = base.drop(base[-1::].index, axis=0)
treino.tail()
```

```python
# Definir 1 para target quando o valor de fechamento de amanh for maior que o de hoje
treino.loc[treino['target'] > treino['Close'], 'target'] = 1
```

```python
# Definir 0 para target quando o valor de fechamento de amanh no for maior
treino.loc[treino['target'] != 1, 'target'] = 0
```

```python
# Separar as variveis dependentes (y) e independentes (x)
y = treino['target']
x = treino.drop('target', axis=1)
```

```python
# Separar os dados de treino e teste
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
# Criar e treinar o modelo de classificao Extra Trees
from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)
```

```python
# Avaliar a acurcia do modelo
resultado = modelo.score(x_teste, y_teste)
print("Acurcia:", resultado)
```

```python
# Fazer a previso para os dados de amanh
prev['target'] = modelo.predict(prev)
```

```python
# Exibir a previso de amanh
print("\nPreviso para amanh:")
print(prev)
```

```python
# Converter as previses em mensagens
def converter_previsao(previsao):
    if previsao == 1:
        return "Invista, a previso  positiva!"
    else:
        return "No invista, a previso  negativa!"
```

```python
# Aplicar a converso das previses
prev['mensagem'] = prev['target'].apply(converter_previsao)
```

```python
# Mostrar a previso final com mensagens
print("\nPrevises com mensagens:")
print(prev[['target', 'mensagem']])
```
